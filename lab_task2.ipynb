```python
# Hard-coded values
length = 12
width = 8

area = length * width
print("Length:", length)
print("Width:", width)
print("Area of Rectangle:", area)
```

```python
# Accepting input from user
student_name = input("Enter student name: ")
marks = int(input("Enter marks: "))

try:
    marks = int(marks)
    if marks >= 50 and marks <= 100:
        result = "Pass"
    elif marks >= 0 and marks < 50:
        result = "Fail"
    else:
      print("Marks should be between 0 and 100.")
      exit()
except ValueError:
    print("Invalid input. Marks should be a number.")
    exit()

print("Name:", student_name)
print("Marks:", marks)
print("Result:", result)
```

```python
# Reading a single value from file
file = open("single_value.txt", "w")
file.write("7")
file.close()

try:
    file = open("single_value.txt", "r")
    number = int(file.read())
    file.close()

    print("Number from file:", number)
    print("Square of number:", number * number)

except FileNotFoundError:
    print("Error: File not found")
except ValueError:
    print("Error: Invalid data in file")
```

```python
# Reading multiple values from file
file = open("numbers.txt", "w")
file.write("10\n20\n30\n40\n50")
file.close()


try:
    file = open("numbers.txt", "r")
    values = file.readlines()
    file.close()

    total = 0
    for val in values:
        total += int(val.strip())

    print("Numbers read:", values)
    print("Sum of numbers:", total)

except FileNotFoundError:
    print("File not found")
except ValueError:
    print("Error: Invalid data in file")
```
